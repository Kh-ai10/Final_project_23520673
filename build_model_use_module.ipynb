```python
#preprocessing "Species" column, mã hóa object thành number
from sklearn.preprocessing import LabelEncoder
dataFrame["Species"] = LabelEncoder().fit_transform(dataFrame["Species"])
print(dataFrame)     #xem kết quả sau khi mã hóa
```

```python
#train_test_split
'''
Chia data thành tập train và tập test
Train: dùng tập này để huấn luyện model, model sẽ đưa ra dự đoán dựa trên tập này
Test: không dùng để train mà dùng để đánh giá hiệu suất model dựa trên dữ liệu mà nó chưa từng nhìn thấy
Có thể thay đổi tỉ trọng train_test trong quá trình huấn luyện model
'''
from sklearn.model_selection import train_test_split
X = dataFrame.drop(columns = ["Species"])    #Lấy tất cả feature trừ cột Species
y = dataFrame["Species"]        #Lấy target Species
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)  #có thể thay đổi test_size thành train_size cũng như thay đổi trọng số
```

```python
#Build decision tree and evaluate accuracy
'''
Ở đây em sẽ thử với cả ba tiêu chuẩn tách là gini, entropy và log-loss
'''
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import  accuracy_score

for criterion in ['gini', 'entropy', 'log_loss']:   #lựa chọn tiêu chuẩn tách
  for max_depth in range(2, 7):   #điều chỉnh tham số
    for min_samples_leaf in range (20, 55 , 5):

        gohomeNT = DecisionTreeClassifier(criterion = criterion, max_depth = max_depth, min_samples_leaf = min_samples_leaf )
        gohomeNT = gohomeNT.fit(X_train,y_train)
        train_pre = gohomeNT.predict(X_train)
        test_pre = gohomeNT.predict(X_test)
        train_acc = accuracy_score(y_train, train_pre)
        test_acc = accuracy_score(y_test, test_pre)

        print( f'criterion:{criterion}, max_depth: {max_depth}, min_samples_leaf: {min_samples_leaf} \ntrain_acc: {train_acc} \ntest_acc: {test_acc} \n') 
```
